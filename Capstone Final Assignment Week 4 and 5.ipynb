```python
# Import required modules
import pandas as pd
import requests as requests
import bs4 as bs4
from geopy.geocoders import Nominatim
import folium
```

```python
# Get data of the neighborhoods from wikipedia
link = 'https://en.wikipedia.org/wiki/List_of_neighbourhoods_in_Bangalore'
```

```python
# Fetch data from site 
# Table 1
table_1 = pd.read_html(link)[0]
table_data_1 = pd.DataFrame(table_1)
table_data_1 = table_data_1.drop(['Image','Summary'], axis =1)

# Table 2
table_2 = pd.read_html(link)[1]
table_data_2 = pd.DataFrame(table_2)
table_data_2 = table_data_2.drop(['Image','Summary'], axis =1)

# Table 3
table_3 = pd.read_html(link)[2]
table_data_3 = pd.DataFrame(table_3)
table_data_3 = table_data_3.drop(['Image','Summary'], axis =1)

# Table 4
table_4 = pd.read_html(link)[3]
table_data_4 = pd.DataFrame(table_4)
table_data_4 = table_data_4.drop(['Image','Summary'], axis =1)

# Table 5
table_5 = pd.read_html(link)[4]
table_data_5 = pd.DataFrame(table_5)
table_data_5 = table_data_5.drop(['Image','Summary'], axis =1)
```

```python
# Concatinating collected data into one dataframe
table_data = pd.concat([table_data_1,table_data_2,table_data_3,table_data_4,table_data_5], axis = 0)
table_data = table_data.reset_index()
table_data = table_data.drop(['index'], axis = 1)
table_data
```

```python
# Prepare name of neighborhoods to catch their latitudde and longtitudes from geolocator package
table_data['Name'] = table_data['Name']+',Bangalore'
neighborhood = table_data['Name'].tolist()
```

```python
# Extract latitude and longtitude for each neighborhood
geolocator = Nominatim(user_agent="mygeocoder")
coordinates_of_neighborhood = []
for i in neighborhood:
    location = geolocator.geocode(i.format(i))
    if location:
        coordinates_of_neighborhood.append([i, location.latitude, location.longitude])
```

```python
# Convert coordinates_of_neighborhood to a data frame
data_frame_of_coordinates = pd.DataFrame(coordinates_of_neighborhood, columns=["Name", "Latitude","Longitude"])
data_frame_of_coordinates
```

```python
# Get the coordinate of Bangalore city to found a better sight about the problem
demand_city = 'Bangalore,India'
geolocator = Nominatim(user_agent="mygeocoder")
location = geolocator.geocode(demand_city)
latitude_of_city = location.latitude
longtitude_of_city = location.longitude
print(f'City: Bangalore, India - latitude: {latitude_of_city} - longtitude: {longtitude_of_city}.')
```

```python
# Location of the city in map
m = folium.Map(location=[latitude_of_city, longtitude_of_city], zoom_start=5)
m
```

```python
# Location of the neighborhoods of Bangalore in map
for latitude, longtitude, neighborhood in zip(data_frame_of_coordinates['Latitude'],  data_frame_of_coordinates['Longitude'], data_frame_of_coordinates['Name']):
    label = f'{neighborhood}'
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker([latitude, longtitude], radius=5, popup=label, color='blue', fill=True, fill_color='#3186cc', fill_opacity=0.7).add_to(m)
m   
```
