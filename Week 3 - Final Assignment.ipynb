# Question 1
---------

```python
# Read data from wikipedia
url='https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M'
data = requests.get(url).text
```

```python
# Extract tables
soup = BeautifulSoup(data,'html.parser')
wiki_table=soup.find('table')
```

```python
# Create dataframe
table_contents=[]
table=soup.find('table')
for row in table.findAll('td'):
    cell = {}
    if row.span.text=='Not assigned':
        pass
    else:
        cell['PostalCode'] = row.p.text[:3]
        cell['Borough'] = (row.span.text).split('(')[0]
        cell['Neighborhood'] = (((((row.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(cell)
```

```python
# Prepare dataframe
df=pd.DataFrame(table_contents)
df['Borough']=df['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
                                             'East TorontoBusiness reply mail Processing Centre969 Eastern':'East Toronto Business',
                                             'EtobicokeNorthwest':'Etobicoke Northwest','East YorkEast Toronto':'East York/East Toronto',
                                             'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})
df.shape
```

```python
# Read Geo data
datastr = requests.get('http://cocl.us/Geospatial_data',allow_redirects=True).text
data_file = io.StringIO(datastr)
df_geo = pd.read_csv(data_file)
```

```python
# Prepare geo data
df_geo.rename(columns={'Postal Code':'PostalCode'},inplace=True)
```

```python
# Merge two data
df_merged = pd.merge(df_geo, df, on='PostalCode')
df_merged =df_merged[['PostalCode','Borough','Neighborhood','Latitude','Longitude']]
df_merged.head(10)
```

```python
# Prepare toronto dataframe
df_toronto = df_merged[df_merged['Borough'].str.contains("Toronto")]
df_toronto.head()
```
